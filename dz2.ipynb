Задание 3

Список поисковых запросов. Получить распределение количества слов в них. Т. е. поисковых запросов из одного слова 5%, из двух - 7%, из трех - 3% итд.
queries = [
‘смотреть сериалы онлайн’,
‘новости спорта’,
‘афиша кино’,
‘курс доллара’,
‘сериалы этим летом’,
‘курс по питону’,
‘сериалы про спорт’,
]

```python
queries = [
'смотреть сери алы онлайн',
'новости спорта',
'афиша кино',
'курс доллара',
'сериалы этим летом',
'курс по питону',
'сериалы про спорт',
]
freq=[0,0,0,0,0,0,0]

# рассмотрим каждый элемент списка с номером k
for k in range(len(queries)):
    i=0
    for letter in queries[k]:
        if letter==' ':
            i=i+1
            pass
    freq[k]=i+1 #количество слов больше на единицу, чем пробелов        
print('Количество слов в каждом списке', freq)
n2=0
n3=0
n4=0
for i in range(len(queries)):
    if freq[i]==2: 
        n2=n2+1
    if freq[i]==3:
        n3=n3+1
    if freq[i]==4:
        n4=n4+1 
#print(n2,n3,n4)                  
print('из двух слов   ', round(100*n2/len(queries)),'%')
print('из трех слов   ', round(100*n3/len(queries)),'%')
print('из четырех слов', round(100*n4/len(queries)),'%')
```
